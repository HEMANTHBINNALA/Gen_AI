```python
# Cell 2: Imports
# Import all required LangChain prompt classes.

from langchain_core.prompts import PromptTemplate, ChatPromptTemplate

print("All imports loaded successfully.")
```

```python
# ============================================================
# Cell 3: Task 1 — Replace Hardcoded Prompts
# Original bad approach:  f"Explain {topic} in simple terms"
# Fixed approach:         PromptTemplate with input_variables
# User provides the topic dynamically via input().
# ============================================================

# --- Define the PromptTemplate (no f-strings, no hardcoded values) ---
explain_topic_prompt = PromptTemplate(
    template="Explain {topic} in simple terms for beginners",
    input_variables=["topic"]
)

def explain_topic(topic):
    """
    Returns a formatted prompt for the given topic.
    Uses PromptTemplate — no f-strings.

    Args:
        topic (str): The topic to explain.

    Returns:
        str: Formatted prompt string.
    """
    return explain_topic_prompt.format(topic=topic)


# --- Dynamic Input ---
print("===== TASK 1: Replace Hardcoded Prompts =====")
user_topic = input("Enter a topic to explain: ").strip()

result = explain_topic(user_topic)
print("\nGenerated Prompt:")
print(result)
```

```python
# ============================================================
# Cell 4: Task 2 — Multi-Input Prompt System
# Template:  "Explain {topic} for {audience} in a {tone} tone"
# All three values are entered dynamically by the user.
# ============================================================

# --- Define the multi-input PromptTemplate ---
multi_input_prompt = PromptTemplate(
    template="Explain {topic} for {audience} in a {tone} tone",
    input_variables=["topic", "audience", "tone"]
)

def generate_multi_input_prompt(topic, audience, tone):
    """
    Formats the multi-input template with provided values.

    Args:
        topic    (str): Subject to explain.
        audience (str): Who the explanation is for.
        tone     (str): The tone to use.

    Returns:
        str: Formatted prompt string.
    """
    return multi_input_prompt.format(topic=topic, audience=audience, tone=tone)


# --- Dynamic Inputs ---
print("===== TASK 2: Multi-Input Prompt System =====")
print("Enter values for the prompt (press Enter after each):")

user_topic    = input("  Topic    : ").strip()
user_audience = input("  Audience : ").strip()
user_tone     = input("  Tone     : ").strip()

result = generate_multi_input_prompt(
    topic=user_topic,
    audience=user_audience,
    tone=user_tone
)

print("\nGenerated Prompt:")
print(result)
```

```python
# ============================================================
# Cell 5: Task 3 — Prompt Variations Engine
# Three PromptTemplates: teaching | interview | storytelling
# User selects which style to use and provides the topic.
# Entering 'all' runs all three templates at once.
# ============================================================

# --- Define 3 style-specific PromptTemplates ---
PROMPT_VARIATIONS = {
    "teaching":     PromptTemplate(template="Explain {topic} clearly step by step",  input_variables=["topic"]),
    "interview":    PromptTemplate(template="Ask 3 questions about {topic}",          input_variables=["topic"]),
    "storytelling": PromptTemplate(template="Explain {topic} as a story",             input_variables=["topic"])
}

def run_variation(topic, style):
    """
    Runs the chosen style template with the given topic.
    Pass style='all' to run every template.

    Args:
        topic (str): The subject to use.
        style (str): 'teaching', 'interview', 'storytelling', or 'all'.

    Returns:
        dict: Mapping of style -> formatted prompt.
    """
    style = style.lower().strip()

    if style == "all":
        # Run every template and return all results
        return {name: tmpl.format(topic=topic) for name, tmpl in PROMPT_VARIATIONS.items()}

    if style not in PROMPT_VARIATIONS:
        raise ValueError(
            "Unknown style '{s}'. Choose from: {valid} or 'all'.".format(
                s=style, valid=list(PROMPT_VARIATIONS.keys())
            )
        )

    return {style: PROMPT_VARIATIONS[style].format(topic=topic)}


# --- Dynamic Inputs ---
print("===== TASK 3: Prompt Variations Engine =====")
print("Available styles: teaching | interview | storytelling | all")

user_topic = input("  Enter topic : ").strip()
user_style = input("  Enter style : ").strip()

try:
    variations = run_variation(topic=user_topic, style=user_style)
    print("\nGenerated Prompt(s):")
    for style_name, prompt_text in variations.items():
        print("  [" + style_name.upper() + "] " + prompt_text)
except ValueError as e:
    print("Error: " + str(e))
```

```python
# ============================================================
# Cell 6: Task 4 — ChatPromptTemplate System
# Roles: teacher | interviewer | motivator
# Each role has a unique system message.
# User enters role + topic dynamically.
# ============================================================

# --- Role-to-SystemMessage behaviour mapping ---
ROLE_BEHAVIORS = {
    "teacher":     "You are a patient and knowledgeable teacher. Explain concepts clearly using simple language and step-by-step breakdowns.",
    "interviewer": "You are a sharp technical interviewer. Ask insightful and relevant questions about the topic to assess depth of knowledge.",
    "motivator":   "You are an energetic and inspiring motivator. Make the topic exciting, build confidence, and encourage curiosity."
}

def build_chat_prompt(role, topic):
    """
    Builds and formats a ChatPromptTemplate for the given role and topic.

    Args:
        role  (str): One of 'teacher', 'interviewer', 'motivator'.
        topic (str): The subject to discuss.

    Returns:
        list: Formatted chat message objects (system + human).

    Raises:
        ValueError: If the role is not recognised.
    """
    role = role.lower().strip()

    if role not in ROLE_BEHAVIORS:
        raise ValueError(
            "Unknown role '{r}'. Choose from: {valid}".format(
                r=role, valid=list(ROLE_BEHAVIORS.keys())
            )
        )

    # Build ChatPromptTemplate: system defines behaviour, human provides the task
    chat_template = ChatPromptTemplate.from_messages([
        ("system", ROLE_BEHAVIORS[role]),
        ("human",  "Please help me understand: {topic}")
    ])

    return chat_template.format_messages(topic=topic)


# --- Dynamic Inputs ---
print("===== TASK 4: ChatPromptTemplate System =====")
print("Available roles: teacher | interviewer | motivator")

user_role  = input("  Enter role  : ").strip()
user_topic = input("  Enter topic : ").strip()

try:
    messages = build_chat_prompt(role=user_role, topic=user_topic)
    print("\nChat Messages Generated:")
    for msg in messages:
        print("  [" + msg.type.upper() + "] " + msg.content)
except ValueError as e:
    print("Error: " + str(e))
```

```python
# ============================================================
# Cell 7: Task 5 — Input Validation Layer
# Validates audience and tone against allowed values.
# User enters values; the function either corrects them
# with defaults (use_defaults=True) or raises a ValueError.
# ============================================================

# Allowed values
VALID_AUDIENCES = ["beginner", "intermediate", "expert"]
VALID_TONES     = ["formal", "casual", "fun"]

# Safe fallback defaults
DEFAULT_AUDIENCE = "beginner"
DEFAULT_TONE     = "formal"

def validate_inputs(audience, tone, use_defaults=True):
    """
    Validates audience and tone against allowed lists.

    Args:
        audience     (str):  Target audience string.
        tone         (str):  Tone string.
        use_defaults (bool): If True, invalid values are replaced with defaults.
                             If False, a ValueError is raised.

    Returns:
        tuple (str, str): (validated_audience, validated_tone)

    Raises:
        ValueError: If use_defaults=False and any input is invalid.
    """
    audience = audience.lower().strip()
    tone     = tone.lower().strip()

    # Validate audience
    if audience not in VALID_AUDIENCES:
        if use_defaults:
            print(
                "[WARNING] '{a}' is not a valid audience. Defaulting to '{d}'.".format(
                    a=audience, d=DEFAULT_AUDIENCE
                )
            )
            audience = DEFAULT_AUDIENCE
        else:
            raise ValueError(
                "Invalid audience '{a}'. Must be one of: {v}".format(
                    a=audience, v=VALID_AUDIENCES
                )
            )

    # Validate tone
    if tone not in VALID_TONES:
        if use_defaults:
            print(
                "[WARNING] '{t}' is not a valid tone. Defaulting to '{d}'.".format(
                    t=tone, d=DEFAULT_TONE
                )
            )
            tone = DEFAULT_TONE
        else:
            raise ValueError(
                "Invalid tone '{t}'. Must be one of: {v}".format(
                    t=tone, v=VALID_TONES
                )
            )

    return audience, tone


# --- Dynamic Inputs ---
print("===== TASK 5: Input Validation Layer =====")
print("Valid audiences : beginner | intermediate | expert")
print("Valid tones     : formal | casual | fun")
print("(Try entering an invalid value to see the validation in action!)")

user_audience = input("\n  Enter audience : ").strip()
user_tone     = input("  Enter tone     : ").strip()

# Ask whether to use defaults or raise an error
mode_choice = input("  Use defaults on invalid input? (yes/no) : ").strip().lower()
use_def = mode_choice != "no"

try:
    validated_audience, validated_tone = validate_inputs(
        audience=user_audience, tone=user_tone, use_defaults=use_def
    )
    print("\nValidation Result:")
    print("  Audience -> " + validated_audience)
    print("  Tone     -> " + validated_tone)
except ValueError as e:
    print("\nValidation Error: " + str(e))
```

```python
# ============================================================
# Cell 8: Task 6 — Prompt Generator App
# Unified app that takes FOUR dynamic inputs:
#   topic | audience | tone | style
# Internally: validates inputs then selects + formats a PromptTemplate.
# ============================================================

# --- Style-specific PromptTemplates (topic + audience + tone) ---
STYLE_TEMPLATES = {
    "teaching": PromptTemplate(
        template="Explain {topic} for {audience} in a {tone} teaching style",
        input_variables=["topic", "audience", "tone"]
    ),
    "interview": PromptTemplate(
        template="Ask 3 interview questions about {topic} suited for a {audience} in a {tone} style",
        input_variables=["topic", "audience", "tone"]
    ),
    "storytelling": PromptTemplate(
        template="Explain {topic} as a {tone} story aimed at a {audience}",
        input_variables=["topic", "audience", "tone"]
    )
}

VALID_STYLES = list(STYLE_TEMPLATES.keys())

def generate_prompt(topic, audience, tone, style):
    """
    Full prompt generation pipeline:
      1. Validate audience & tone
      2. Validate style
      3. Select the matching PromptTemplate
      4. Return the formatted prompt

    Args:
        topic    (str): The subject.
        audience (str): Target audience (beginner/intermediate/expert).
        tone     (str): Tone (formal/casual/fun).
        style    (str): Output style (teaching/interview/storytelling).

    Returns:
        str: Formatted prompt string.
    """
    # Step 1: Validate audience and tone (auto-correct with defaults)
    validated_audience, validated_tone = validate_inputs(
        audience=audience, tone=tone, use_defaults=True
    )

    # Step 2: Validate style
    style = style.lower().strip()
    if style not in VALID_STYLES:
        raise ValueError(
            "Invalid style '{s}'. Choose from: {v}".format(s=style, v=VALID_STYLES)
        )

    # Step 3: Select and format the template
    return STYLE_TEMPLATES[style].format(
        topic=topic,
        audience=validated_audience,
        tone=validated_tone
    )


# --- Dynamic Inputs ---
print("===== TASK 6: Prompt Generator App =====")
print("Styles    : teaching | interview | storytelling")
print("Audiences : beginner | intermediate | expert")
print("Tones     : formal | casual | fun")
print()

user_topic    = input("  Topic    : ").strip()
user_audience = input("  Audience : ").strip()
user_tone     = input("  Tone     : ").strip()
user_style    = input("  Style    : ").strip()

try:
    final_prompt = generate_prompt(
        topic=user_topic,
        audience=user_audience,
        tone=user_tone,
        style=user_style
    )
    print("\nGenerated Prompt:")
    print(final_prompt)
except ValueError as e:
    print("\nError: " + str(e))
```

```python
# ============================================================
# Cell 9: Task 7 — Template Reusability Test
# ONE PromptTemplate. User enters N sets of inputs.
# Demonstrates: same structure -> different outputs each time.
# ============================================================

# --- Single reusable PromptTemplate ---
reusable_prompt = PromptTemplate(
    template="Explain {topic} to a {audience} in a {tone} way",
    input_variables=["topic", "audience", "tone"]
)

def run_reusable_template(inputs_list):
    """
    Formats ONE PromptTemplate against a list of input dictionaries.

    Args:
        inputs_list (list of dict): Each dict must have 'topic', 'audience', 'tone'.

    Returns:
        list of str: All formatted prompts.
    """
    results = []
    for inp in inputs_list:
        formatted = reusable_prompt.format(
            topic=inp["topic"],
            audience=inp["audience"],
            tone=inp["tone"]
        )
        results.append(formatted)
    return results


# --- Dynamic Inputs: user decides how many sets to enter ---
print("===== TASK 7: Template Reusability Test =====")

try:
    num_inputs = int(input("How many input sets do you want to test? (e.g. 3): ").strip())
except ValueError:
    print("Invalid number. Defaulting to 3 sets.")
    num_inputs = 3

user_inputs = []
for i in range(num_inputs):
    print("\n  -- Input Set " + str(i + 1) + " --")
    t = input("    Topic    : ").strip()
    a = input("    Audience : ").strip()
    n = input("    Tone     : ").strip()
    user_inputs.append({"topic": t, "audience": a, "tone": n})

print("\nOne Template -> " + str(num_inputs) + " Different Outputs:")
print("-" * 55)

outputs = run_reusable_template(user_inputs)
for idx, out in enumerate(outputs):
    print("Output " + str(idx + 1) + " : " + out)
```

```python
# ============================================================
# Cell 10: Full Pipeline Demo
# Combines ALL tasks into one interactive session.
# User provides ALL inputs dynamically.
# Pipeline: Input -> Validation -> PromptTemplate -> ChatPromptTemplate -> Output
# ============================================================

def full_pipeline(topic, audience, tone, style, role):
    """
    Full end-to-end prompt generation pipeline.

    Steps:
      1. Receive raw user inputs
      2. Validate audience + tone
      3. Generate styled prompt via PromptTemplate
      4. Build role-based ChatPromptTemplate messages
      5. Return and display all outputs

    Args:
        topic    (str): Subject to cover.
        audience (str): Target audience.
        tone     (str): Explanation tone.
        style    (str): Prompt style (teaching/interview/storytelling).
        role     (str): ChatPromptTemplate role (teacher/interviewer/motivator).

    Returns:
        dict: {'prompt': str, 'chat_messages': list}
    """
    separator = "=" * 60

    print(separator)
    print("           FULL PIPELINE EXECUTION")
    print(separator)

    # Step 1 — Display raw inputs
    print("[Step 1] Raw Inputs Received:")
    print("  Topic    : " + topic)
    print("  Audience : " + audience)
    print("  Tone     : " + tone)
    print("  Style    : " + style)
    print("  Role     : " + role)

    # Step 2 — Validate
    print("\n[Step 2] Validating inputs...")
    validated_audience, validated_tone = validate_inputs(
        audience=audience, tone=tone, use_defaults=True
    )
    print("  Audience (validated) : " + validated_audience)
    print("  Tone     (validated) : " + validated_tone)

    # Step 3 — Generate styled prompt
    print("\n[Step 3] Generating styled prompt...")
    generated_prompt = generate_prompt(
        topic=topic,
        audience=validated_audience,
        tone=validated_tone,
        style=style
    )
    print("  Styled Prompt: " + generated_prompt)

    # Step 4 — Build role-based ChatPromptTemplate
    print("\n[Step 4] Building ChatPromptTemplate for role: '" + role + "'...")
    chat_messages = build_chat_prompt(role=role, topic=topic)
    for msg in chat_messages:
        print("  [" + msg.type.upper() + "]: " + msg.content)

    # Step 5 — Final output
    print("\n[Step 5] Pipeline complete.")
    print(separator)

    return {"prompt": generated_prompt, "chat_messages": chat_messages}


# --- Dynamic Inputs for Full Pipeline ---
print("===== FULL PIPELINE: All Tasks Combined =====")
print("Styles    : teaching | interview | storytelling")
print("Audiences : beginner | intermediate | expert")
print("Tones     : formal | casual | fun")
print("Roles     : teacher | interviewer | motivator")
print()

pipe_topic    = input("  Topic    : ").strip()
pipe_audience = input("  Audience : ").strip()
pipe_tone     = input("  Tone     : ").strip()
pipe_style    = input("  Style    : ").strip()
pipe_role     = input("  Role     : ").strip()

try:
    pipeline_output = full_pipeline(
        topic=pipe_topic,
        audience=pipe_audience,
        tone=pipe_tone,
        style=pipe_style,
        role=pipe_role
    )
    print("\nFinal Styled Prompt:")
    print(">> " + pipeline_output["prompt"])
except ValueError as e:
    print("\nPipeline Error: " + str(e))
```
